```python
from matplotlib.mathtext import np
import numpy
import cv2
import matplotlib.pyplot as plt
import glob
import numpy as np
```

```python
ds_estudiantes=[]
ds_estudiantes_labels=[]
```

```python
x = "/content/0 - copia (2).png"
test_image = cv2.imread(x, cv2.IMREAD_GRAYSCALE )
```

```python
plt.imshow(test_image, cmap="gray")
```

```python
img_resized = cv2.resize(test_image, (28,28), interpolation = cv2.INTER_LINEAR)
img_resized = cv2.bitwise_not(img_resized)
```

```python
plt.imshow(img_resized, cmap="gray")
```

```python
ds_estudiantes.append(img_resized)
ds_estudiantes_labels.append("0")
```
